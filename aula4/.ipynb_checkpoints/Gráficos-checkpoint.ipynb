```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_and_create_array(file_dir):
    array = []
    f = open(file_dir)
    for line in f:
        if len(line) <= 5:
            array.append(int(line))
    array = np.array(array)
    return array
```

```python
cp = read_and_create_array('entrega_CP.txt')
cpd = read_and_create_array('entrega_CPD.txt')
lr  = read_and_create_array('entrega_left_right.txt')
lfr = read_and_create_array('entrega_left_forward_right.txt')
```

## Gráficos

```python
plt.plot(lr)
plt.title('Controle Left-Right')
plt.ylabel('valor do sensor de luz')
plt.xlabel('tempo em amostras')
```

```python
plt.plot(lfr)
plt.title('Controle Left-Forward-Right')
plt.ylabel('valor do sensor de luz')
plt.xlabel('tempo em amostras')
```

```python
plt.plot(cp-45)
plt.title('Controle Proporcional (KP = 5)')
plt.ylabel('erro')
plt.xlabel('tempo em amostras')
```

```python
plt.plot(cpd-45)
plt.title('Controle Proporcional Derivativo (KP = 3.5 | KD = 1.0)')
plt.ylabel('erro')
plt.xlabel('tempo em amostras')
```
